```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
import numpy as np

from sgd_feedforward import SgdNeuralNetwork
from scg_feedforward import ScgNeuralNetwork
from leapfrog_feedforward import LeapfrogNeuralNetwork
import network_optimizer
```

## Wine

```python
# Read the data
df = pd.read_csv('./data/winequality-red.csv')

# Define features (X) and labels (Y)
X = df.iloc[:, :-1].values  # All columns except the last one
Y = df.iloc[:, -1].values  # The last column

# Standardize the features
scalers = [StandardScaler() for _ in range(X.shape[1])]
scaler_Y = StandardScaler()
X = np.hstack([scaler.fit_transform(X[:, i].reshape(-1, 1)) for i, scaler in enumerate(scalers)])
Y = scaler_Y.fit_transform(Y.reshape(-1, 1)).flatten()



# Split the data into training and testing sets (80%/20%)
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
num_hidden_units = network_optimizer.optimize_sgd_number_of_units(train_X, train_Y, 11, 3, 1)
optimum_epochs, train_performance, valid_performance = network_optimizer.optimize_sgd_epochs(train_X, train_Y, 11, num_hidden_units, 1)

print('Optimum hidden units:', num_hidden_units)
print('Optimum epochs:', optimum_epochs)
network_optimizer.graph_classifier_performance(train_performance, valid_performance, 'SGD MSE on Wine')

train_avg, train_std, test_avg, test_std = network_optimizer.evaluate_sgd(11, num_hidden_units, 1, train_X, train_Y, test_X, test_Y, num_runs=3)

print(train_avg, train_std, test_avg, test_std)
```

```python
num_hidden_units = network_optimizer.optimize_scg_number_of_units(train_X, train_Y, 11, 3, 1)
optimum_epochs, train_performance, valid_performance = network_optimizer.optimize_scg_epochs(train_X, train_Y, 11, num_hidden_units, 1)

print('Optimum hidden units:', num_hidden_units)
print('Optimum epochs:', optimum_epochs)
network_optimizer.graph_classifier_performance(train_performance, valid_performance, 'MSE accuracy on Wine')

train_avg, train_std, test_avg, test_std = network_optimizer.evaluate_scg(11, num_hidden_units, 1, train_X, train_Y, test_X, test_Y, num_runs=3)

print(train_avg, train_std, test_avg, test_std)
```

```python
num_hidden_units = network_optimizer.optimize_leapfrog_number_of_units(train_X, train_Y, 11, 3, 1)
optimum_epochs, train_performance, valid_performance = network_optimizer.optimize_leapfrog_epochs(train_X, train_Y, 11, num_hidden_units, 1)

print('Optimum hidden units:', num_hidden_units)
print('Optimum epochs:', optimum_epochs)
network_optimizer.graph_classifier_performance(train_performance, valid_performance, 'Leapfrog MSE on Wine')

train_avg, train_std, test_avg, test_std = network_optimizer.evaluate_leapfrog(11, num_hidden_units, 1, train_X, train_Y, test_X, test_Y, num_runs=3)

print(train_avg, train_std, test_avg, test_std)
```
